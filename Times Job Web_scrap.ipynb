```python
from bs4 import BeautifulSoup
import requests as re 
import pandas as pd 
import numpy as np
```

```python
link='https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=Data+Analyst&txtLocation='
```

```python
response=re.get(link).text
```

```python
soup=BeautifulSoup(response,'html.parser')
```

```python
print(soup)
```

```python
box=soup.find_all('li',class_="clearfix job-bx wht-shd-bx")
```

# Job_title 

```python
job_title=[]
for i in box:
    job_title.append(i.find('h2').text.replace('\n',"").replace('\r',"").strip())
```

```python
job_title
```

```python
i.find('h2').text.replace('\n',"")
```

```python
df=pd.DataFrame({})
```

```python
df['job_title']=job_title
```
